```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./train.csv/train.csv')
df.sample(10)
```

```python
df = df.sample(30000)
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
que_df = df[['question1','question2']]
que_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer

questions = list(que_df['question1']) + list(que_df['question2'])

cv = CountVectorizer(max_features=3000)

q1_arr, q2_arr = np.vsplit(cv.fit_transform(questions).toarray(),2)
```

```python
q1_temp = pd.DataFrame(q1_arr, index=que_df.index)
q2_temp = pd.DataFrame(q2_arr, index=que_df.index)
```

```python
final_que_df = pd.concat([q1_temp,q2_temp],axis=1)
final_que_df
```

```python
final_que_df['is_duplicate'] = df['is_duplicate']
```

```python
final_que_df.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(final_que_df.iloc[:,0:-1].values, final_que_df.iloc[:,-1].values, test_size=0.2, random_state=5)
```

```python


from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
y_pred = rf.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
```

```python
plt.figure(figsize=(3,4))
sns.heatmap(cm,annot=True,fmt='d')
plt.show()
plt.xlabel('Predicted')
plt.Y
```
